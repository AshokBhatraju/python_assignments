```python
import pandas as pd
```

```python
#loading the data
df2016 = pd.read_csv('UIC2016Basketball.csv')
df2016.head()
```

```python
#Assign column headers
df2016 = pd.read_csv('UIC2016Basketball.csv')

df2016.columns =['Date','Opponent','UIC_Score','Opp_Score','UIC Field Goal Percentage', 'Opp Field Goal Percentage','UIC 3 point Field Goal Percentage', 'Opp 3 point Field Goal Percentage', 'UIC Rebound', 'Opp Rebound', 'UIC Assists', 'Opp Assists']

df2016.head()
```

```python
#Fill the missing data cell with a hyphen sympol ‘-‘. 
df2016.fillna('-')
```

```python
#What are the data types used in this data frame? 
df2016.dtypes
```

```python
#How many school opponents did Men’s Basketball played in 2016?
opp= df2016.groupby(['Opponent']).count()
for row in opp.index:
    print(row, end="\n")
print("Number of school opponents: ", len(opp.index))
```

```python
#List all the games that UIC scored more than 65 points
UIC65=df2016.query('UIC_Score > 65')
UIC65[['Date','Opponent', 'UIC_Score']]
```

```python
 # Calculate numbers of games of Wins, Losses, and Ties

win=df2016.query('UIC_Score > Opp_Score')
loss=df2016.query('UIC_Score < Opp_Score')
tie= df2016.query('UIC_Score == Opp_Score')

print("WINS")
print(win[['Opponent', 'UIC_Score', 'Opp_Score']])
print("")
print("Number of WINS: ", len(win.index))

print("---------------------------------------------")
print("LOSSES")
print(loss[['Opponent', 'UIC_Score', 'Opp_Score']])
print("")
print("Number of LOSSES: ", len(loss.index))


print("---------------------------------------------")
print("TIES")
print(tie[['Opponent', 'UIC_Score', 'Opp_Score']])
print("")
print("Number of TIES: ", len(tie.index))
```
